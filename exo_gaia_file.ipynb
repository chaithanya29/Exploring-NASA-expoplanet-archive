```python
def write_part_rows(df, ini, end, name):
    df1 = df.iloc[ini:end]
    df1.to_csv("/home/chaithanya/kepler_names/"+name+".csv", index=False, header=False)
```

```python
def split_write(df):
    i=0
    ln = len(df)
    for k in range(int(ln/100) +1):
        if (i/100 == int(ln/100)):
            write_part_rows(df, i, i+1+(ln%100), str(i)+"_to_"+str(ln%100))
        else:
            write_part_rows(df, i, i+100, str(i)+"_to_"+str(i+100))
        i += 100
    
```

```python
split_write(df1)
```
